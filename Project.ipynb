```python
config_file="ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
frozen_model="frozen_inference_graph.pb"
```

```python
model=cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
model.setInputSize(120,120)
model.setInputScale(1.0/127.5)
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

```python
import cv2

cap = cv2.VideoCapture("5834623-uhd_3840_2160_24fps.mp4")

if not cap.isOpened():
    cap = cv2.VideoCapture(0)  # Fall back to webcam if video file fails
if not cap.isOpened():
    raise IOError("Cannot open video")

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

# Define target size for resizing (width x height)
resize_width = 800
resize_height = 450

while True:
    ret, frame = cap.read()
    
    # Validate if the frame is successfully read
    if not ret or frame is None:
        print("Failed to read frame or end of video. Exiting...")
        break

    # Resize the frame to the target size
    frame = cv2.resize(frame, (resize_width, resize_height))

    # Perform object detection
    try:
        ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)
    except Exception as e:
        print(f"Error during detection: {e}")
        break

    print(ClassIndex)

    # Check if any object is detected
    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <= 80:  # Assuming 80 is the maximum class index
                cv2.rectangle(frame, boxes, (255, 0, 0), 2)
                cv2.putText(frame, classLabels[ClassInd - 1], 
                            (boxes[0] + 10, boxes[1] + 40), 
                            font, fontScale=font_scale, 
                            color=(0, 255, 0), thickness=3)

    # Show the resized frame with detection
    cv2.imshow("Object Detection Project", frame)

    # Quit if 'q' is pressed
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
